```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset for Data Analytics - Sheet1.csv")

# ==========================
# 1. Check Missing Values
# ==========================
print("Missing Values:")
print(df.isnull().sum())

# Fill missing CouponCode values
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python

# ==========================
# 2. Remove Duplicate Rows
# ==========================
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
# ==========================
# 3. Correct Data Formats
# ==========================

# Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Convert categorical columns
df["PaymentMethod"] = df["PaymentMethod"].astype("category")
df["OrderStatus"] = df["OrderStatus"].astype("category")
df["ReferralSource"] = df["ReferralSource"].astype("category")

# Ensure numeric columns are numeric
df["Quantity"] = pd.to_numeric(df["Quantity"])
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"])
df["ItemsInCart"] = pd.to_numeric(df["ItemsInCart"])
df["TotalPrice"] = pd.to_numeric(df["TotalPrice"])
```

```python
# ==========================
# Final Information
# ==========================
print("\nDataset Info:")
print(df.info())

print("\nFirst Five Rows:")
print(df.head())

# Save cleaned dataset
df.to_csv("Cleaned_Dataset.csv", index=False)

print("\nCleaned dataset saved as 'Cleaned_Dataset.csv'")
```
