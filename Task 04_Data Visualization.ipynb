```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Cleaned_Dataset.csv")

df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Visualization 1: Monthly Sales Trend (Line Chart)
df["Month"] = df["Date"].dt.to_period("M").astype(str)

monthly_sales = df.groupby("Month")["TotalPrice"].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
# Visualization 2: Top 10 Selling Products (Bar Chart)

top_products = df.groupby("Product")["Quantity"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.bar(top_products.index, top_products.values)

plt.xticks(rotation=45)

plt.title("Top 10 Selling Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")

plt.show()
```

```python
# Visualization 3: Payment Method Distribution (Pie Chart)

payment = df["PaymentMethod"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(payment,
        labels=payment.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Payment Method Distribution")

plt.show()
```

```python
# Visualization 4: Order Status (Bar Chart)

status = df["OrderStatus"].value_counts()

plt.figure(figsize=(7,5))

plt.bar(status.index, status.values)

plt.title("Order Status")
plt.xlabel("Status")
plt.ylabel("Number of Orders")

plt.show()
```

```python
# Visualization 5: Quantity Distribution (Histogram)

plt.figure(figsize=(8,5))

plt.hist(df["Quantity"], bins=10)

plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

```python
# Visualization 6: Total Price Distribution (Histogram)

plt.figure(figsize=(8,5))

plt.hist(df["TotalPrice"], bins=20)

plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")

plt.show()
```

```python
# Visualization 7: Outliers (Box Plot)

plt.figure(figsize=(7,5))

plt.boxplot(df["TotalPrice"])

plt.title("Box Plot of Total Price")

plt.show()
```

```python
# Visualization 8: Sales by Referral Source (Bar Chart)

referral = df.groupby("ReferralSource")["TotalPrice"].sum()

plt.figure(figsize=(8,5))

plt.bar(referral.index, referral.values)

plt.xticks(rotation=45)

plt.title("Sales by Referral Source")
plt.xlabel("Referral Source")
plt.ylabel("Revenue")

plt.show()
```

```python
# Visualization 9: Average Sales by Payment Method

avg_sales = df.groupby("PaymentMethod")["TotalPrice"].mean()

plt.figure(figsize=(8,5))

plt.bar(avg_sales.index, avg_sales.values)

plt.title("Average Order Value by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Average Total Price")

plt.xticks(rotation=45)

plt.show()
```

```python
# Visualization 10: Correlation Heatmap

corr = df[["Quantity","UnitPrice","ItemsInCart","TotalPrice"]].corr()

plt.figure(figsize=(6,5))

plt.imshow(corr)

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
